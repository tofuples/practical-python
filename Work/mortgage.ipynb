How much will Dave pay if he pays an extra $1000/month for 4 years starting after the first five years have already been paid?

```python
total = 500000.0 #valor do emprestimo
taxa_anual = 0.05 
pagamento_mensal = 2684.11
total_pago = 0.0
total_meses = 0

extra = 1000
extra_inicio = 61
extra_fim = 108

while total > 0: #loop continua até o total a ser pago zerar
    total_meses = total_meses + 1 
    taxa_mensal = total * (taxa_anual/12) #valor que vai pra pagar o juros (o total deve ser atualizado todo mês)
    total = total + taxa_mensal  #o valor total a ser pago passa a ser total + juros do primeiro mês
    
    if total_meses >= extra_inicio and total_meses <= extra_fim:
        total = total - extra
        total_pago = total_pago + extra
        
    if total < pagamento_mensal:
        pagamento_mensal = total

    total = total - pagamento_mensal # atualiza o valor total descontando o pagamento fixo que será feito no primeiro mês
    #total = total + taxa_mensal - pagamento_mensal
    
    total_pago = total_pago + pagamento_mensal 
    
    print(total_meses, round(total_pago,2), total)
    
print(f'O total pago será ${round(total_pago,2)} ao longo de {total_meses}')


```
